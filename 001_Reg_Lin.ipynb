```python
import numpy as np
import pylab as py
```

```python
#Tworzenie ciagu uczacego
n = 11
x = np.linspace(0, 10, n)
X = np.ones((n, 2))
X[:,1] = x
theta = np.array([1,3]).reshape(2,1)
Y = np.dot(X,theta) + 3*np.random.randn(n, 1)
py.plot(X[:,1], Y, 'bo')
py.show()
```

```python
#Algorytm rownan normalnych
def alg_r_norm(X,Y):
  theta = np.dot( np.dot( np.linalg.inv( np.dot(X.T,X) ),X.T ),Y )
  return theta
```

```python
theta_est = alg_r_norm(X,Y)
print("ROWNANIA NORMALNE\n")
print("Prawdziwe wartosci parametrow: ", theta.T)
print("Estymowane wartosci parametrow: ", theta_est.T)
py.plot(X[:,1], Y, 'bo')
py.plot(X[:,1], np.dot(X,theta_est), 'g')
py.xlabel("x")
py.ylabel("y")
py.title("Regresja z rownan normalnych")
py.show()
```

```python
#Algorytm gradientowy stochastyczny
def alg_grad_stoch( X, Y, alpha = 0.001, theta0 = np.array([0,0]).reshape(2,1) ):
  py.plot(X[:,1], np.dot(X,theta0), 'r')
  for i in range (500):
    ind = np.random.randint(n)
    theta0 = theta0 - alpha * ( np.dot(X[ind],theta0) - Y[ind] ) * X[ind,1]
    py.plot(X[:,1], np.dot(X,theta0), 'r')
  return theta0
```

```python
theta_est = alg_grad_stoch(X,Y)
print("ALGORYTM GRADIENTOWY STOCHASTYCZNY\n")
print("Prawdziwe wartosci parametrow: ", theta.T)
print("Estymowane wartosci parametrow: ", theta_est.T)
py.plot(X[:,1], Y, 'bo') 
py.plot(X[:,1], np.dot(X,theta_est), 'g')
py.xlabel("x") 
py.ylabel("y")
py.title("Regresja z algorytmu gradientowego stochastycznego")
py.show()
```

```python
#Algorytm gradientowy zbiorczy
def alg_grad_zb( X, Y, alpha = 0.001, theta0 = np.array([0,0]).reshape(2,1) ):
  py.plot(X[:,1], np.dot(X,theta0), 'r')
  for i in range(25):
    theta0 = theta0 - alpha * np.dot( X.T, np.dot(X,theta0 ) - Y)
    py.plot( X[:,1], np.dot(X,theta0), 'r')
  return theta0
```

```python
theta_est = alg_grad_zb(X,Y)
print("ALGORYTM GRADIENTOWY ZBIORCZY\n")
print("Prawdziwe wartosci parametrow: ", theta.T)
print("Estymowane wartosci parametrow: ", theta_est.T)
py.plot(X[:,1], Y, 'bo')
py.plot(X[:,1], np.dot(X,theta_est), 'g')
py.xlabel("x")
py.ylabel("y")
py.title("Regresja z algorytmu gradientowego zbiorczego")
py.show()
```
